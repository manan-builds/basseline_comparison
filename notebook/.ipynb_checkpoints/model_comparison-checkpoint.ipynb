```python
from transformers import pipeline
from datasets import load_dataset
from sklearn.metrics import accuracy_score
```

```python
dataset=load_dataset("imdb")
test_data = dataset["test"].select(range(200))
```

```python
models = {
"distilbert_sst2": "distilbert-base-uncased-finetuned-sst-2-english",
"twitter_roberta": "cardiffnlp/twitter-roberta-base-sentiment"
}
```

```python
for name, model_id in models.items():
    clf = pipeline("sentiment-analysis", model=model_id)
    y_true = []
    y_pred = []
```

```python
for item in test_data:
    text = item["text"][:512]
    true_label = item["label"]
    pred = clf(text)[0]["label"]
if "POS" in pred.upper():
    pred_label = 1
else:
    pred_label = 0
    
    y_true.append(true_label)
    y_pred.append(pred_label)
acc = accuracy_score(y_true, y_pred)
print(name, "accuracy:", acc)
```

```python
def predict_with_threshold(text, clf, threshold=0.80):
    out = clf(text)[0]
    label = out["label"]
    score = out["score"]
if score < threshold:
    return "UNCERTAIN", score
return label, score
```
