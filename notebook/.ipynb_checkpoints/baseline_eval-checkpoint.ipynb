```python
import csv
```

```python
def save_baseline_results(dataset_split, clf, filename="baseline_results.csv", n=100):
    sample = dataset_split.select(range(n))

    with open(filename, "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(["text_preview", "true_label", "pred_label", "confidence"])

        for item in sample:
            text = item["text"][:200]
            true_label = item["label"]
            out = clf(item["text"][:512])[0]
            pred_label = out["label"]
            confidence = out["score"]
            
            writer.writerow([text, true_label, pred_label, confidence])
```
